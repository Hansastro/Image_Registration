# Draw some text on an image and its bounding box with PIL

```python
from PIL import Image, ImageFont, ImageDraw, ImageEnhance
from matplotlib import pyplot as plt
```

```python
file_name = r'./Test_Images/Form/template.png'
out_file = r'./out.png'
```

```python
font_file = r'./fonts/nimbus-sans-l/NimbusSanL-Reg.otf'
```

```python
text_xy=(20, 225)
```

```python
text = "0123456789"
source_img = Image.open(file_name).convert("RGBA")
font_size = 40
font = ImageFont.truetype(font_file, font_size)

draw = ImageDraw.Draw(source_img)
#draw.rectangle(((0, 0), (300, 100)), fill="black")
draw.text(text_xy, text, font=font, fill ="black")   # ttf for PIL >= 1.1.4

text_size = font.getsize(text)
shape = [text_xy, (text_size[0]+text_xy[0], text_size[1]+text_xy[1])]
draw.rectangle(shape, outline="#47c459", width=2)

source_img.save(out_file, "PNG")
```
